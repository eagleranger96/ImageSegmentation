```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import cv2
```

```python
def getSegmentedImage(img, colors):
    X = img.reshape(-1,3)
    
    model = KMeans(colors)
    model.fit(X)
    
    color_palette = model.cluster_centers_
    
    gen_img = np.zeros([X.shape[0],3])
    for i in range(len(gen_img)):
        gen_img[i] = color_palette[model.labels_[i]]
    
    gen_img = gen_img.reshape(img.shape)
    
    return gen_img
```

```python
def printImages(images, columns):
    plt.figure(figsize=(20,10))
    for i, image in enumerate(images):
        plt.subplot(len(images) / columns + 1, columns, i + 1)
        plt.imshow(image)
```

```python
cat_images = []
cat_images.append(cv2.cvtColor(cv2.imread("Images/cat.jpg"), cv2.COLOR_BGR2RGB))
cat_images.append(getSegmentedImage(cat_images[0],7)/255)
cat_images.append(getSegmentedImage(cat_images[0],5)/255)
cat_images.append(getSegmentedImage(cat_images[0],3)/255)
printImages(cat_images,4)
```
